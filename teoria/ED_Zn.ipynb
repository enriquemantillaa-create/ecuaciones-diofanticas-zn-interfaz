# $\text{Propiedades en }\mathbb{Z}$

## $\text{Algoritmo de la división}$

Sea $a,b\in \mathbb{Z}$ existen enteros $q,r \in \mathbb{Z}$ con $0 \leq r < b$ tales que 
$$a=b\cdot q+r$$

```python
def algoritmo_division(a,b):
    if (b==0):
        return 0, a
    if (a<0 and (int(a/b)*b != a) and b<0):
        q= int(a/b) + 1
    elif (a<0 and (int(a/b)*b != a) and b>0):
        q= int(a/b) - 1
    else:
        q = int(a/b)
    r = a - q*b
    return q,r
```

## $\text{Algoritmo de Euclides}$

Sea $a,b \in \mathbb{Z}$ aplicamos el algoritmo de la división hasta que nuestro ultimo residuo sea $0$, de la siguiente forma:
$$a=b\cdot q_1 + r_1$$
$$b=r_1\cdot q_2 + r_2$$
$$r_1=r_2 \cdot q_3 + r_3$$
$$\vdots$$
$$r_{n-1}=r_{n} \cdot q_{n+1} + r_{n+1}$$
$$r_{n} = r_{n+1}\cdot q_{n+2} + 0$$

donde $r_{n+1}$ es el maximo común divisor de $a$ y $b$.

```python
def algoritmo_euclides(a, b):
    cocientes = []
    residuos =[]
    q, r = algoritmo_division(a, b)
    if (r==0):
        cocientes.append(q)
        residuos.append(r)
    else:
        while (r != 0):
            cocientes.append(q)
            residuos.append(r)
            a = b
            b = r
            q, r = algoritmo_division(a, b)
    return b, cocientes, residuos
```

## $\text{Maximo Comun Divisor}$

Sea $a,b \in \mathbb{Z}$ no ambos cero, su maximo común divisor es el entero $d$ que satisface:
- $d | a$ y $d | b$
- Si $d' | a$ y $d' | b$ entonces $d' \leq d$ (d es el maximo comun divisor).

Para encontar el maximo comun divisor de dos enteros $a$ y $b$ se puede usar el algoritmo de Euclides.

```python
def mcd(a,b):
    d, c, r = algoritmo_euclides(a,b)
    return d
```

## $\text{Lema de Bezout}$

Sea $a,b \in \mathbb{Z}$ existen enteros $x,y \in \mathbb{Z}$ tales que $$ax+by=mcd(a,b)$$

Estos enteros se pueden calcular mediante el algoritmo de Euclides, de la siguiente forma:

Sea $a,b \in \mathbb{Z}$ por el algoritmo de Euclides tenemos que:
$$\begin{array}{c|c}
    a = b\cdot q_1 + r_1 & r_1= a - b\cdot q_1 \\
    b = r_1 \cdot  q_2 + r_2 & r_2 = b - r_1 \cdot q_2\\
    r_1 = r_2 \cdot  q_3 + r_3 & r_3 = r_1 - r_2 \cdot q_3\\
    \vdots & \vdots\\
    r_{k-1} = r_{k} \cdot  q_{k+1} + r_{k+1} & r_{k+1} = r_{k-1} - r_{k} \cdot q_{k+1}
\end{array}$$
Donde el $mcd(a,b)=r_{k+1}$. Ahora hacemos sustitucion regresiva sobre este algoritmo, como primer paso tenemos que 
$$r_{k+1} =  0 \cdot 0 + r_{k+1}\cdot 1  \longrightarrow x= 0,y = 1$$
$$r_{k+1} =  0 \cdot x + r_{k+1}\cdot y$$

Ahora sustituimos $r_{k+1}$, entonces
$$r_{k+1} =  0 \cdot 0 + (r_{k-1} - r_{k} \cdot q_{k+1})\cdot 1$$
$$r_{k+1} = r_{k-1}\cdot 1 - r_{k} \cdot q_{k+1}\cdot 1$$
$$r_{k+1} = r_{k-1}\cdot y + r_{k} ( x - q_{k+1}\cdot y) \longrightarrow y= ( x - q_{k+1}\cdot y),x = y$$
$$r_{k+1} = r_{k-1}\cdot x + r_{k}\cdot y $$

Ahora si sustituimos $r_{k} = r_{k-2}-r{k-1}\cdot q_{k}$, tenemos:
$$r_{k+1} = r_{k-1}\cdot x + (r_{k-2}-r{k-1}\cdot q_{k}) y $$
$$r_{k+1} = r_{k-1} (x - q_{k}\cdot y) + r_{k-2}\cdot y \longrightarrow y= ( x - q_{k}\cdot y),x = y$$
$$r_{k+1} = r_{k-1}\cdot y + r_{k-2}\cdot x $$

Notemos que:
$$r_{k+1} = r_{k-1}\cdot x + r_{k}\cdot y $$
$$r_{k+1} = r_{k-2}\cdot x + r_{k-1}\cdot y$$
$$r_{k+1} = r_{k-3}\cdot x + r_{k-2}\cdot y$$
$$\vdots$$
$$r_{k+1} = a\cdot x + b\cdot y$$
con $x= y$ y $y= x - q\cdot y$, donde $q=q_{k+1},q_{k},...,q_1$.

```python
def lema_bezout(a,b):
   if (a==0):
        return b, 0, 1
   d, cocientes, residuos = algoritmo_euclides(a,b)
   if (residuos[0] == 0):
      x = 0
      y = 1
   else:
      x = 0
      y = 1
      for q in reversed(cocientes):
         aux = y
         y = x - q*y
         x = aux
   return d, x, y
```

# $\text{Propiedades en } \mathbb{Z}_n$

## $\text{Algoritmo de la división en } \mathbb{Z}_n$

Sea $\overline{a}, \overline{b} \in \mathbb{Z}_n$, existen clases $\overline{q}, \overline{r} \in \mathbb{Z}_n$ tales que

$$\overline{a} = \overline{b}\cdot\overline{q} + \overline{r}$$

Para encontrar $\overline{q}, \overline{r}$ debemos recordar que para $\overline{a}, \overline{b} \in \mathbb{Z}_n$, decimos que $\overline {b}$ divide a $\overline{a}$ si existe $\overline{q} \in \mathbb{Z}_n$ tal que 
$$\overline{a}=\overline{b} \cdot \overline{q}$$
entonces 
$$a \equiv b\cdot q \pmod{n}$$
esto equivale a la ecuacion diofantica de la forma
$$b \cdot q + n \cdot y = a$$
la cual sabemos que tiene solución si  y solo si $mcd(b,n)$ divide a $a$, es decir que $b | a$ si y solo si $mcd(b,n) | a $.

Dado lo anterior con el $\text{lema de Bezout}$ podemos encontrar $u,v\in \mathbb{Z}$ tales que 
$$b\cdot u + n\cdot v = mcd(b,n)$$
llamemos $d=mcd(b,n)$, por el algoritmo de la division tenemos que existen $k,r \in \mathbb{Z}$ con $0\leq r <d$ tales que
$$a= d\cdot k + r$$
Ahora
$$b\cdot u + n\cdot v = d$$
$$b\cdot u \cdot k + n\cdot v \cdot k + r= d\cdot k + r$$
Si llamamos $q=u \cdot k$ y $y= v \cdot k$, entonces 
$$b\cdot q + n\cdot y + r= a$$
esto implica que 
$$\overline{b}\cdot \overline{q} + \overline{r} = \overline{a} $$
Si $\overline{r}=0$ implica que $\overline{b} | \overline{a}$ 


```python
def algoritmo_division_zn(a,b,n):
    d, u, v = lema_bezout(b,n)
    k,r = algoritmo_division(a,d)
    q= u*k
    q1, q = algoritmo_division(q, n)
    return q, r
```

## $\text{Algoritmo de Euclides en } \mathbb{Z}_n$

Sea $\overline{a}, \overline{b} \in \mathbb{Z}_n$ aplicamos el algoritmo de la división en $\mathbb{Z}_n$ analogamente al de los numeros enteros hasta encontrar $\overline{r_{k+1}}= \overline{0}$, al final como ultimo paso si queremos encontrar con ayuda de este algoritmo el maxímo común en $\mathbb{Z}_n$ calculamos el maxímo común divisor de $r_{k+1}$ y $n$ en $\mathbb{Z}$.

```python
def algoritmo_euclides_zn(a,b,n):
    cocientes = []
    residuos = []
    q, r = algoritmo_division_zn(a, b, n)
    if (r== 0):
        cocientes.append(q)
        residuos.append(r)
    else:
        while r != 0:
            cocientes.append(q)
            residuos.append(r)
            a = b
            b = r
            q, r = algoritmo_division_zn(a, b, n)
    mcd,c, r =algoritmo_euclides(b,n)

    return mcd, cocientes, residuos
```

```python
print(algoritmo_euclides_zn(2,3,6))
```

entonces:
$$\overline{2} = \overline{3}\cdot \overline{0} + \overline{2}$$
$$\overline{3} = \overline{2}\cdot \overline{1} + \overline{1}$$
$$\overline{2} = \overline{1}\cdot \overline{2} + \overline{0}$$

por lo que $MCD(\overline{2}, \overline{3})= \overline{mcd(2,3,1)} = \overline{1}$ y $\overline{3}$ no divide a $\overline{2}$, ahora si cambiamos intencionalmente nuestro modulo a $n=5$, tenemos que:

por lo que $MCD(\overline{2}, \overline{3}) = \overline{mcd(2,3,5)}=1$ y en este caso $\overline{3}$ si divide a $\overline{2}$.

## $\text{Maximo común divisor en } \mathbb{Z}_n$

Sea $\overline{a}, \overline{b} \in \mathbb{Z}_n$ entonces su maximo común divisor esta defino como:
$$MCD(\overline{a},\overline{b})=\overline{mcd(a,b,n)}$$

```python
def MCD (a,b,n):
    d,c,r= algoritmo_euclides_zn(a,b,n)
    return d
```

## $\text{Lema de Bezout en } \mathbb{Z}_n$

Sea $\overline{a}, \overline{b} \in \mathbb{Z}_n$, entonces existen clases 
$\overline{x}, \overline{y} \in \mathbb{Z}_n$ tales que:
$$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = MCD(\overline{a},\overline{b})$$


Si llamamos $\overline{d}= MCD(\overline{a},\overline{b})$, tenemos que:
$$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{d}$$
esta ecuacion es equivalente a la congruencia
$$a\cdot x + b\cdot y \equiv d \pmod{n}$$
lo que nos lleva a la ecuación diofantica lineal de la forma
$$ax+ by + nk = d \text{ con } k \in \mathbb{Z}$$
la cual tiene solución si y solo si $mcd(a,b,n)$ divide a $d$, en este caso como son iguales, por la propiedad $4.2.1$ ($\overline{x} | \overline{x}$) esto siempre se cumple.

Para encontrar $\overline{x}$ y $\overline{y}$ usamos el Lema de Bezout en $\mathbb{Z}$ de la siguiente manera:

Para $a,b$ existen enteros $u,v \in \mathbb{Z}$ tales que $$au+bv=d_1$$ 
donde $d_1= mcd(a,b)$. Ahora para los enteros $d_1, n$ existen enteros $u_1, v_1 \in \mathbb{Z}$ tales que $$d_1u_1+nv_1=d$$ sustiyuyendo $d_1$, tenemos que $$(au+bv)u_1+ nv_1 =d$$
$$a(uu_1) + b(vu_1) + nv_1 =d$$
por lo tanto $x= uu_1$ y $y= vu_1$

```python
def lema_bezout_zn(a,b,n):
    d1, u ,v = lema_bezout(a,b)
    d, u1, v1= lema_bezout(d1,n)
    x= u*u1
    y= v*u1
    q1,x= algoritmo_division(x,n)
    q2,y= algoritmo_division(y,n)
    return d, x, y
```

$\textbf{Ejemplo}$: Sea $\overline{25}, \overline{55} \in  \mathbb{Z}_{100}$ por lema de Bezout en $\mathbb{Z}_{n}$, tenemos que:

```python
print(lema_bezout_zn(25, 55, 100))
```

- Como solución obtuvimos que $\overline{x}= \overline{98}$ y $\overline{y}=\overline{1}$, cumpliendo que:
$$\overline{25}\cdot \overline{98} + \overline{55}\cdot \overline{1} = \overline{5}$$
$$\overline{2505} = \overline{5}$$
$$\overline{5} = \overline{5}$$

# $\text{Metodos de Solución } (\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{c})$

## $\text{¿Existe solución?}$

$\textbf{Teorema 4.3.1.}$ Sea la ecuación diofantica lineal $$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{c},$$
con $\overline{a}, \overline{b}, \overline{c} \in \mathbb{Z}_{n}$, la cual tiene solución si y solo si $mcd(a,b,n)$ divide a $c$.

```python
def hay_solucion(a, b, c, n):
    solucion = False
    d = MCD(a,b,n)
    q,r = algoritmo_division(c,d)
    if r == 0:
        solucion = True
    return solucion
    
```

## $\text{Falsa Posición}$

$\textbf{Teorema 4.3.2.}$ Sea la ecuación diofántica de la forma $$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{c},$$ con $\overline{a},\overline{b},\overline{c} \in \mathbb{Z}_n$. Sean $\overline{x_1},\overline{y_1} \in \mathbb{Z}_n$ dos clases arbitrarias tales que $$\overline{a}\cdot \overline{x_1} + \overline{b}\cdot \overline{y_1}= \overline{d},$$ si $\overline{d}$ es unidad entonces existe un factor $\overline{k} = \overline{c} \cdot \overline{d}^{-1}$ y las clases $\overline{x},\overline{y}$ estan definidas por $\overline{x} = \overline{x_1} \cdot \overline{k}$ y $\overline{y} = \overline{y_1} \cdot \overline{k}$ son solución de la ecuación original en $\mathbb{Z}_n$.

De lo contrario si $\overline{d}$ no es unidad pero el $mcd(d,n)$ divide a $c$ sigue existiendo algún factor $k \in \mathbb{Z}_n$ tal que $\overline{x} = \overline{x_1} \cdot \overline{k}$ y $\overline{y} = \overline{y_1} \cdot \overline{k}$ son soluciones
de la ecuación original en $\mathbb{Z}_n$. 

```python
def falsa_posicion(a,b,c,n,x1,y1):
    s = hay_solucion(a,b,c,n)
    if s == False:
        return "No hay solucion para la ecuacion diofantica lineal en Zn"
    d= a*x1 + b*y1
    q0,d= algoritmo_division(d,n)
    m = mcd(d,n)
    q,r= algoritmo_division(c,d)
    if (m==1):
        m, d_i , l = lema_bezout(d,n)
        k= c*d_i
        x= x1*k
        y= y1*k
        q1, x = algoritmo_division(x,n)
        q2, y = algoritmo_division(y,n)
    elif (m!=1 and r==0):
        x= x1*q
        y= y1*q
        q1, x = algoritmo_division(x,n)
        q2, y = algoritmo_division(y,n)
    else:
        return "x1, y1, no nos ayuda a resolver la ecuacion diofantica lineal"

    return x ,y
```

$\text{Ejemplo}$: Sea la ecuación diofántica en $\mathbb{Z}_{12}$
$$\overline{5}\cdot \overline{x} + \overline{7}\cdot \overline{y} = \overline{4}$$

tomemos $\overline{x_1}= \overline{6}$ y $\overline{y_1}= \overline{8}$

```python
print(falsa_posicion(5,7,4,12,6,8))
```

- Nuestro resultado fue $\overline{x}= \overline{0}$, y $\overline{y}= \overline{4}$, lo cual cumple
$$\overline{5}\cdot \overline{0} + \overline{7}\cdot \overline{4} = \overline{28} = \overline{4}$$

## $\text{Algoritmo de Euclides y Lema de Bezout}$

Sea la ecuación diofántica de la forma $$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{c},$$
Si tiene solucion en $\mathbb{Z}_n$ significa que $mcd(a,b,n)$ divide $c$. Ahora por el Lema de Bezout en $\mathbb{Z}_n$ existen clases $\overline{u},\overline{v} \in \mathbb{Z}_n$ tales que:
 $$\overline{a}\cdot \overline{u} + \overline{b}\cdot \overline{v} = \overline{mcd(a,b,n)},$$
llamemos $\overline{d}= \overline{mcd(a,b,n)}$ como el $mcd(a,b,n)$ divide $c$ entonces $\overline{d}$ divide a $\overline{c}$ por lo tanto existe un $\overline{k} \in \mathbb{Z}_n$ tal que $$\overline{c}= \overline{d}\cdot \overline{k} $$
Por lo que basta multiplicar la ecuacion del lema de bezout por $\overline{k}$, es decir
$$\overline{a}\cdot \overline{u}\cdot \overline{k} + \overline{b}\cdot \overline{v} \cdot \overline{k} = \overline{d}\cdot \overline{k}$$
si llamamos $\overline{x}= \overline{u} \cdot \overline{k}$ y $\overline{y}= \overline{v} \cdot \overline{k}$, tenemos que:
$$\overline{a}\cdot \overline{x} + \overline{b}\cdot \overline{y} = \overline{c}$$

```python
def euclides_bezout(a,b,c,n):
    s = hay_solucion(a,b,c,n)
    if s == False:
        return "No hay solucion para la ecuacion diofantica lineal en Zn"
    d, u, v = lema_bezout_zn(a,b,n)
    k, r = algoritmo_division(c, d)
    x= u*k
    y= v*k
    q,x= algoritmo_division(x, n)
    q,y= algoritmo_division(y, n)
    return x, y
```

$\textbf{Ejemplo}$: Sea la ecuación diofantica lineal en $\mathbb{Z}_6$
$$\overline{3}\cdot \overline{x} + \overline{2}\cdot \overline{y} = \overline{5}$$

```python
print(euclides_bezout(3,2,5,6))
```

Por lo tanto $\overline{x}= \overline{5}$ y $\overline{y}= \overline{1}$, cumpliendo que:
$$\overline{3}\cdot \overline{5} + \overline{2}\cdot \overline{1} = \overline{17} = \overline{5}$$

## $\text{Diofanto}$

Dada la ecuación diofántica en $\mathbb{Z}_n$
$$\overline{a}\cdot \overline{x} + \overline{c} = \overline{b}\cdot \overline{y}$$
Si $d= mcd(a,b,n)$ tal que $d$ divide a $c$, el método de Diofanto es algoritmicamente análogo a como se hace en los números enteros, es decir:
$$\overline{b} = \overline{a}\cdot\overline{q} + \overline{r} \text{, entonces, } \overline{x}= \overline{q}\cdot \overline{y} + \overline{t},$$
$$\overline{a} = \overline{r}\cdot\overline{q_1} + \overline{r_1} \text{, entonces, } \overline{y}= \overline{q_1}\cdot \overline{t} + \overline{t_1},$$
$$\overline{r} = \overline{r_1}\cdot\overline{q_2} + \overline{r_2} \text{, entonces, } \overline{t}= \overline{q_2}\cdot \overline{t_1} + \overline{t_2},$$
$$ \vdots $$
$$ \overline{r_k} = \overline{r_{k+1}}\cdot\overline{q_{k+2}} + \overline{1} \text{, entonces, } \overline{t_k}= \overline{q_{k+2}}\cdot \overline{t_{k+1}} + \overline{t_{k+2}}.$$
Luego se obtiene 
$$\overline{t_{k+1}} + \overline{c} = \overline{r_{k+1}}\cdot \overline{t_{k+2}}$$
Para asegurar que nuestro método sea análogo a $\mathbb{Z}$, se requiere que nuestro ultimo residuo sea la unidad $\overline{1}$.

```python
def diofanto (a,b,c,n,t_2):
    s = hay_solucion(a,b,c,n)
    if s == False:
        return "No hay solucion para la ecuacion diofantica lineal en Zn"
    m, cocientes, residuos = algoritmo_euclides_zn(b,a,n)
    if (residuos[-1] != 1):
        return "El Metodo de Diofanto no es aplicable"
    p=[]
    if (len(residuos) == 1):
        y = a*t_2 + c
        x= t_2 + cocientes[0]*y 
        q, x = algoritmo_division(x,n)
        q1, y = algoritmo_division(y,n)
    else:
        t_1 = residuos[-2]*t_2 + c
        p.append(t_2)
        p.append(t_1)
        for q in reversed(cocientes):
            t = q*t_1 + t_2
            p.append(t)
            aux = t_1
            t_1= t
            t_2= aux
        x= p[-1]
        y= p[-2]
        q, x = algoritmo_division(x,n)
        q1, y = algoritmo_division(y,n)
    return x,y
```

$\textbf{Ejemplo}$: Sea la ecuación diofantica en $\mathbb{Z}_{14}$:
$$\overline{8}\cdot \overline{x} + \overline{2} = \overline{5}\cdot \overline{y}$$


```python
print(diofanto(8,5,2,14,3))
```

- Por lo que $\overline{x}= \overline{9}$ y $\overline{y}= \overline{12}$, cumpliendo que  
$$\overline{8}\cdot \overline{9} + \overline{2} = \overline{5}\cdot \overline{12}$$
$$\overline{74} = \overline{60}$$
$$\overline{4} = \overline{4}$$

# $\text{Ternas Pitagoricas en } \mathbb{Z}_n (\overline{x}^2 + \overline{y}^2 = \overline{z}^2)$

## $\text{Metodo de Diofanto}$

Para las igualdades logradas por Diofanto:
$$(m^2 + 1)^2 = (2m)^2 + (m^2-1)^2 \text { para } m\in \mathbb{Z}$$
$$(p^2 + q^2)^2 = (2pq)^2 + (p^2-q^2)^2 \text { para } p,q\in \mathbb{Z} \text{ tales que } p>q.$$
Basta aplicar la clase de equivalencia módulo $n$, es decir:
$$(\overline{m}^2 + \overline{1})^2 = (\overline{2}\cdot \overline{m})^2 + (\overline{m}^2-\overline{1})^2 \text { para } \overline{m}\in \mathbb{Z}_n$$
$$(\overline{p}^2 + \overline{q}^2)^2 = (\overline{2}\cdot \overline{p}\cdot \overline{q})^2 + (\overline{p}^2-\overline{q}^2)^2 \text { para } \overline{p},\overline{q}\in \mathbb{Z}_n $$

```python
def ternas_Diofanto1 (m,n):
    z= m**2+1
    x= 2*m
    y= m**2 -1
    q1,x = algoritmo_division(x,n)
    q2,y = algoritmo_division(y,n)
    q3,z = algoritmo_division(z,n)
    return x,y,z
```

```python
print(ternas_Diofanto1(8,6))
```

```python
def ternas_Diofanto2 (p,q,n):
    z= p**2 + q**2
    x= 2*p*q
    y= p**2 - q**2
    q1,x = algoritmo_division(x,n)
    q2,y = algoritmo_division(y,n)
    q3,z = algoritmo_division(z,n)
    return x,y,z
```

```python
print(ternas_Diofanto2(7,2,11))
```

## $\text{Metodo de Fibonacci}$

Analogo al metodo de Diofanto tenemos que:
$$\overline{x}= \overline{f_m}\cdot \overline{f_{m+3}}$$
$$\overline{y}= \overline{2}\cdot \overline{f_{m+1}}\cdot \overline{f_{m+2}}$$
$$\overline{z}= \overline{f_{m+1}}^2 + \overline{f_{m+2}}^2$$

Donde $f_{m}, f_{m+1}, f_{m+2}, f_{m+3}$ son elementos consecutivos de la sucesión de Fibonacci.

```python
def sucesion_fibonacci(n):
  f1 = 1
  if n<=f1:
    return n
  else:
    fn = sucesion_fibonacci(n-1) + sucesion_fibonacci(n-2)
    return fn
```
